### Uni-variate Analysis

```python
# Dropping the unique identifier column
print("shape before dropping:",data.shape)
data.drop('Id', axis=1, inplace=True)
print("shape before dropping:",data.shape)
```

```python
# Check outliers for 'MSSubClass' column
plt.boxplot(data['MSSubClass'])
sns.title('MSSubClass')
sns.show()
```

##### MSSubClass has extreme positive outliers and needs outlier treatment

```python
# Check the distribution of 'MSSubClass' using either Histogram or Distplot
plt.distplot(data['MSSubClass'])
sns.title('MSSubClass')
sns.show()
```

##### Inference: MSSubClass is not following a Normal Distrbution

```python
# Check outliers for 'LotFrontage'
plt.boxplot(data['LotFrontage'])
sns.title('LotFrontage')
sns.show()
```

##### Inference: There are positive and negative outliers in LotFrontage column and needs outliers treatment

```python
# Checking the distribution of LotFrontage using either Histogram or distplot
sns.hist(data['LotFrontage'])
sns.title('LotFrontage')
sns.show()
```

##### Inference: LotFrontage is not following normal distribution

```python
# Check the cardinality of 'Street'
data['Street'].value_counts()/len(data)*100
```

##### Inference: Street column is highly imbalanced

```python
# Check the count of distinct labels and the labels names using nunique & unique functions
print("Number of unique labels in Street Column: ",data['Street'].nunique())
print("Labels in Street Column: ",data['Street'].unique())
```

```python
# Pie chart for Street
sns.pie(data['Street'].value_counts(), autopct='%1.2f%%', labels= ['Pave', 'Grvl'])
sns.title('Street')
sns.show()
```

```python
# Bar chart for Street
data['Street'].value_counts().plot(kind='bar')
sns.title('Street')
sns.show()
```

```python
# Cardinality checking LotShape
data['LotShape'].value_counts()/len(data)*100
```

##### Inference: LotShape has data imbalance issues

```python
sns.figure(figsize=(8,5))
sns.suptitle('Lot Shape', size=30)
sns.subplot(1,2,1)
sns.pie(data['LotShape'].value_counts(), autopct='%1.2f%%', labels= ['Reg', 'IR1', 'IR2', 'IR3'])

sns.subplot(1,2,2)
data['LotShape'].value_counts().plot.bar()
sns.show()
```

## Bi-variate Analysis

```python
# One categorical & One numberical
plt.boxplot(x = 'KitchenQual', y = 'SalePrice', data = data)
sns.title('SalePrice vs KitchenQual')
sns.show()
```

##### Inference: Good and Excellent KithcenQuality houses has higher value compared to Poor/Fair/Average Kitchen Quality houses

```python
# Which type of Foundation has what type of Sales value
plt.boxplot(x = 'Foundation', y = 'SalePrice', data = data)
sns.title('SalePrice vs Foundation')
sns.show()
```
